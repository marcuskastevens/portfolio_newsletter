```python
path = r'chat_gpt_prompts.pickle'
```

```python
with open(path, 'rb') as reader:
    prompts = pickle.load(reader)
prompts
```

```python
for ticker, prompt in prompts.items():
    print('='*50)
    print(f'{ticker} ChatGPT Prompt:')
    print('='*50)
    print(prompt)
```
